# Assignment: Real vs. Fake Dataset Profiling and Comparison

This notebook completes the assignment by:
1. Downloading a real dataset from data.gov (~1000 rows after filtering).
2. Profiling it with descriptive statistics and charts, saving to Excel.
3. Generating an initial fake dataset mimicking the real one.
4. Profiling the fake dataset and comparing it to the real one.
5. Improving the fake dataset and re-comparing.
6. Analyzing improvements for better simulation of real data.

**Dataset**: U.S. County Population Estimates by Age, Sex (filtered to 2020-2022, Male/Female, first 1000 records).  
**Source**: https://seer.cancer.gov/popdata/yr1990_2023.20ages/us.1990_2023.20ages.adjusted.txt.gz (via data.gov).  
**Columns Used**: 'Year', 'CountyFIPS', 'Sex' (mapped), 'Age' (numeric), 'Population'.

```python
# Import necessary libraries for data handling, visualization, and stats
import pandas as pd  # For data manipulation and reading/writing CSV/Excel
import numpy as np   # For numerical operations and random sampling
import matplotlib.pyplot as plt  # For creating charts
import seaborn as sns  # For enhanced visualizations
from scipy import stats  # For fitting distributions (used in improved fake data)
import warnings  # To suppress warnings for cleaner output
import os  # For file and directory operations
from datetime import datetime  # For timestamped filenames
warnings.filterwarnings('ignore')

# Set style for plots
sns.set_style("whitegrid")  # Makes plots look professional
print("Libraries imported successfully.")  # Verification: Confirm imports work
```

## Step 1: Select and Download Real Dataset
- Dataset: U.S. County population estimates by year, age group, and sex (1990-2023, filtered to 2020-2022).
- Direct .txt.gz URL from SEER/Census (fixed-width format, read with read_fwf).
- Filter to Male/Female sex and 2020-2022 years, then first 1000 records (~1000 rows).
- Save as 'real_dataset.csv' with error handling for permission issues.

```python
# Step 1: Download the real dataset using pandas from the direct URL (fixed-width format)
url = "https://seer.cancer.gov/popdata/yr1990_2023.20ages/us.1990_2023.20ages.adjusted.txt.gz"  # Direct .txt.gz link from SEER/Census
colspecs = [(0,4), (4,6), (6,8), (8,11), (11,12), (12,13), (13,14), (14,16), (16,24)]  # Fixed widths based on dictionary (total 24 bytes)
names = ['Year', 'State', 'StateFIPS', 'CountyFIPS', 'Race', 'Origin', 'Sex', 'Age', 'Population']  # Column names from dictionary
real_df = pd.read_fwf(url, compression='gzip', colspecs=colspecs, names=names, header=None)  # Read fixed-width gzipped file

# Select relevant columns
relevant_columns = ['Year', 'CountyFIPS', 'Sex', 'Age', 'Population']  # Year, County ID, Sex, Age Group, Population
real_df = real_df[relevant_columns]  # Subset to these columns

# Clean data: Convert to numeric where needed, drop NaNs
real_df['Year'] = pd.to_numeric(real_df['Year'], errors='coerce')  # Ensure Year is numeric
real_df['Sex'] = pd.to_numeric(real_df['Sex'], errors='coerce')  # Sex: 1=Male, 2=Female
real_df['Age'] = pd.to_numeric(real_df['Age'], errors='coerce')  # Age: numeric groups
real_df['Population'] = pd.to_numeric(real_df['Population'], errors='coerce')  # Population: numeric
real_df = real_df.dropna(subset=['Year', 'Sex', 'Age', 'Population'])  # Drop invalid rows

# Filter to recent years and sexes (1=Male, 2=Female)
real_df = real_df[real_df['Year'].isin([2020, 2021, 2022])]  # Filter to 2020-2022
real_df = real_df[real_df['Sex'].isin([1, 2])]  # Filter to Male/Female

# Map Sex to strings for better readability
real_df['Sex'] = real_df['Sex'].map({1: 'Male', 2: 'Female'})  # Map numeric to labels

# Ensure Population is int
real_df['Population'] = real_df['Population'].astype(int)

# Filter to first 1000 records if larger
if len(real_df) > 1000:
    real_df = real_df.head(1000)  # Slice to first 1000 rows
    print("Dataset filtered to first 1000 records.")  # Confirmation message

# Display basic info for verification
print("Shape of real dataset (rows, columns):", real_df.shape)  # Shows dimensions; expect ~1000 rows, 5 columns
print("\nColumn names:", real_df.columns.tolist())  # Lists columns: ['Year', 'CountyFIPS', 'Sex', 'Age', 'Population']
print("\nData types:\n", real_df.dtypes)  # Shows types (e.g., Year: int, Population: int)
print("\nFirst 5 rows:\n", real_df.head())  # Displays sample data
print("\nUnique values in each column:")
for col in real_df.columns:
    print(f"{col}: {real_df[col].unique()[:5]}")  # Shows first 5 unique values per column

# Save the real dataset to CSV with error handling
output_dir = "./output"  # Define output directory
if not os.path.exists(output_dir):  # Create directory if it doesn't exist
    os.makedirs(output_dir)
output_path = os.path.join(output_dir, 'real_dataset.csv')  # Use output directory
try:
    real_df.to_csv(output_path, index=False)  # Export without index
    print(f"\nReal dataset saved as '{output_path}'")  # Verification
except PermissionError as e:
    print(f"PermissionError: {e}. Try closing any programs using 'real_dataset.csv' or use a different filename.")
    # Fallback: Use timestamped filename
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    fallback_path = os.path.join(output_dir, f'real_dataset_{timestamp}.csv')
    real_df.to_csv(fallback_path, index=False)
    print(f"Fallback: Saved as '{fallback_path}'")
except Exception as e:
    print(f"Error saving file: {e}. Please check directory permissions or specify another path.")
```

## Step 2: Create Profile for Real Dataset
- Compute descriptive statistics using pandas.describe().
- Generate charts: Bar plot of average population by age group/sex, line plot over years (total trend).
- Save stats and chart data to 'profile_real.xlsx'.
- Charts are shown inline; data saved in Excel for reproducibility.

```python
# Step 2: Generate descriptive statistics for numeric columns
real_stats = real_df.describe(include='all')  # Computes mean, std, min/max, quartiles, etc.
print("Descriptive Statistics for Real Dataset:\n", real_stats)  # Display for verification

# Add mode for categorical columns
real_stats.loc['mode'] = real_df.mode().iloc[0] if not real_df.mode().empty else 'N/A'  # Add mode row

# Save stats to Excel with error handling
output_path = os.path.join(output_dir, 'profile_real.xlsx')
try:
    with pd.ExcelWriter(output_path, mode='w') as writer:  # Write mode
        real_stats.to_excel(writer, sheet_name='Descriptive_Stats')  # Sheet 1: Basic stats
    print(f"Stats saved to '{output_path}'")  # Verification
except PermissionError as e:
    print(f"PermissionError: {e}. Try closing any programs using 'profile_real.xlsx'.")
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    fallback_path = os.path.join(output_dir, f'profile_real_{timestamp}.xlsx')
    with pd.ExcelWriter(fallback_path, mode='w') as writer:
        real_stats.to_excel(writer, sheet_name='Descriptive_Stats')
    output_path = fallback_path
```

```python
# Step 2: Create and save charts for real dataset
fig, axes = plt.subplots(1, 2, figsize=(15, 6))  # Create subplot figure with 2 panels

# Chart 1: Bar plot of average population by age group and sex
pivot_data = real_df.pivot_table(values='Population', index='Age', columns='Sex', aggfunc='mean')  # Pivot to average pop
pivot_data.plot(kind='bar', ax=axes[0])  # Bar plot
axes[0].set_title('Average Population by Age Group and Sex (Real Data)')  # Title
axes[0].set_ylabel('Average Population')  # Y-axis label
axes[0].tick_params(axis='x', rotation=45)  # Rotate x-labels

# Save pivot data to Excel
try:
    with pd.ExcelWriter(output_path, mode='a', engine='openpyxl') as writer:  # Append mode
        pivot_data.to_excel(writer, sheet_name='Chart1_Data')  # Sheet 2
except PermissionError as e:
    print(f"PermissionError: {e}. Using fallback filename.")
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    fallback_path = os.path.join(output_dir, f'profile_real_{timestamp}.xlsx')
    with pd.ExcelWriter(fallback_path, mode='a', engine='openpyxl') as writer:
        pivot_data.to_excel(writer, sheet_name='Chart1_Data')
    output_path = fallback_path

# Chart 2: Line plot of total population over years
yearly_total = real_df.groupby('Year')['Population'].sum()  # Sum population per year
yearly_total.plot(kind='line', ax=axes[1], marker='o')  # Line plot
axes[1].set_title('Total Population Trend Over Years (Real Data)')  # Title
axes[1].set_ylabel('Total Population')  # Y-axis label

# Save yearly data to Excel
try:
    with pd.ExcelWriter(output_path, mode='a', engine='openpyxl') as writer:
        yearly_total.to_excel(writer, sheet_name='Chart2_Data')  # Sheet 3
except PermissionError as e:
    print(f"PermissionError: {e}. Using fallback filename.")
    fallback_path = os.path.join(output_dir, f'profile_real_{timestamp}.xlsx')
    with pd.ExcelWriter(fallback_path, mode='a', engine='openpyxl') as writer:
        yearly_total.to_excel(writer, sheet_name='Chart2_Data')
    output_path = fallback_path

plt.tight_layout()  # Adjust layout
plt.show()  # Display inline
print(f"Charts generated and data saved to '{output_path}'")  # Verification
```

## Step 3: Create Initial Test (Fake) Dataset
- Replicate fields: Year (2020-2022), Age (numeric groups), Sex ('Male'/'Female'), Population (int).
- Generate 500 records (under 1000) to simulate.
- Simple replication: Sample years uniformly, age groups/sexes from real frequencies, population from uniform range based on real min/max.
- Save as 'fake_v1.csv'.

```python
# Step 3: Create initial fake dataset mimicking real structure
np.random.seed(42)  # Set seed for reproducibility

n_records = 500  # Number of fake records (under 1000)

# Extract unique values and ranges from real data
real_years = sorted(real_df['Year'].unique())  # List of unique years
real_ages = real_df['Age'].unique()  # Unique age groups
real_sexes = real_df['Sex'].unique()  # Unique sexes
pop_min, pop_max = real_df['Population'].min(), real_df['Population'].max()  # Min/max for population

# Generate fake data
fake_data = {
    'Year': np.random.choice(real_years, n_records),  # Randomly sample years
    'CountyFIPS': np.random.choice(real_df['CountyFIPS'].unique(), n_records),  # Sample county IDs
    'Sex': np.random.choice(real_sexes, n_records),  # Randomly sample sexes
    'Age': np.random.choice(real_ages, n_records),  # Randomly sample age groups
    'Population': np.random.randint(pop_min, pop_max + 1, n_records)  # Uniform random ints
}

fake_v1_df = pd.DataFrame(fake_data)  # Create DataFrame
output_path = os.path.join(output_dir, 'fake_v1.csv')
try:
    fake_v1_df.to_csv(output_path, index=False)  # Save to CSV
    print(f"Initial fake dataset saved as '{output_path}' with shape:", fake_v1_df.shape)  # Verification
    print(fake_v1_df.head())  # Preview
except PermissionError as e:
    print(f"PermissionError: {e}. Using fallback filename.")
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    fallback_path = os.path.join(output_dir, f'fake_v1_{timestamp}.csv')
    fake_v1_df.to_csv(fallback_path, index=False)
    print(f"Fallback: Saved as '{fallback_path}'")
```

## Steps 4 & 5: Profile Fake v1 and Compare to Real
- Repeat profiling on fake_v1 (stats and charts).
- Comparison: Side-by-side stats and visual overlays on charts.
- Save profile to 'profile_fake_v1.xlsx'.
- Analysis: Initial fake uses uniform sampling, so distributions may not match (e.g., population variance too high).

```python
# Step 4: Profile for fake_v1
fake_v1_stats = fake_v1_df.describe(include='all')  # Descriptive stats
print("Descriptive Statistics for Fake v1:\n", fake_v1_stats)  # Display

# Save to Excel
output_path = os.path.join(output_dir, 'profile_fake_v1.xlsx')
try:
    with pd.ExcelWriter(output_path, mode='w') as writer:
        fake_v1_stats.to_excel(writer, sheet_name='Descriptive_Stats')  # Sheet 1
except PermissionError as e:
    print(f"PermissionError: {e}. Using fallback filename.")
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    fallback_path = os.path.join(output_dir, f'profile_fake_v1_{timestamp}.xlsx')
    with pd.ExcelWriter(fallback_path, mode='w') as writer:
        fake_v1_stats.to_excel(writer, sheet_name='Descriptive_Stats')
    output_path = fallback_path

# Charts for fake_v1
fig, axes = plt.subplots(1, 2, figsize=(15, 6))  # Subplots

# Bar plot for fake
fake_pivot = fake_v1_df.pivot_table(values='Population', index='Age', columns='Sex', aggfunc='mean')  # Pivot
fake_pivot.plot(kind='bar', ax=axes[0], alpha=0.7, label='Fake')  # Bar with transparency
try:
    with pd.ExcelWriter(output_path, mode='a', engine='openpyxl') as writer:
        fake_pivot.to_excel(writer, sheet_name='Chart1_Data')  # Save data
except PermissionError as e:
    print(f"PermissionError: {e}. Using fallback filename.")
    fallback_path = os.path.join(output_dir, f'profile_fake_v1_{timestamp}.xlsx')
    with pd.ExcelWriter(fallback_path, mode='a', engine='openpyxl') as writer:
        fake_pivot.to_excel(writer, sheet_name='Chart1_Data')
    output_path = fallback_path

# Line plot for fake
fake_yearly = fake_v1_df.groupby('Year')['Population'].sum()  # Yearly sum
fake_yearly.plot(kind='line', ax=axes[1], marker='s', label='Fake')  # Line with squares
try:
    with pd.ExcelWriter(output_path, mode='a', engine='openpyxl') as writer:
        fake_yearly.to_excel(writer, sheet_name='Chart2_Data')  # Save data
except PermissionError as e:
    print(f"PermissionError: {e}. Using fallback filename.")
    fallback_path = os.path.join(output_dir, f'profile_fake_v1_{timestamp}.xlsx')
    with pd.ExcelWriter(fallback_path, mode='a', engine='openpyxl') as writer:
        fake_yearly.to_excel(writer, sheet_name='Chart2_Data')
    output_path = fallback_path

plt.tight_layout()
plt.show()
print(f"Fake v1 profile saved to '{output_path}'")  # Verification
```

```python
# Step 5: Compare profiles visually
fig, axes = plt.subplots(1, 2, figsize=(15, 6))  # Subplots for comparison

# Compare bar plots
pivot_data.plot(kind='bar', ax=axes[0], alpha=0.7, label='Real')  # Real bars
fake_pivot.plot(kind='bar', ax=axes[0], alpha=0.5, label='Fake v1')  # Overlay fake
axes[0].set_title('Comparison: Avg Population by Age/Sex')
axes[0].legend()  # Add legend

# Compare line plots
yearly_total.plot(kind='line', ax=axes[1], marker='o', label='Real')  # Real line
fake_yearly.plot(kind='line', ax=axes[1], marker='s', label='Fake v1')  # Overlay fake
axes[1].set_title('Comparison: Total Population Over Years')
axes[1].legend()

plt.tight_layout()
plt.show()  # Display comparison charts

# Numerical comparison: Compute % difference in means
mean_diff = {}  # Dict for differences
for col in ['Year', 'Population']:  # Compare numeric cols
    real_mean = real_df[col].mean()
    fake_mean = fake_v1_df[col].mean()
    mean_diff[col] = abs((fake_mean - real_mean) / real_mean * 100)  # % absolute difference
print("Mean % Differences (Real vs Fake v1):", mean_diff)  # Print for analysis
```

## Step 6: Improve Accuracy of Fake Dataset
- **Analysis**: Initial fake (v1) uses uniform sampling, leading to unrealistic distributions (e.g., population variance too high).
- **Improvements**:
  - Sample years, ages, and sexes with weights based on real data frequencies.
  - For population: Fit normal distribution per age/sex combo and sample from it.
  - Ensure proportional representation matching real data.
- Generate v2 with 500 records, profile, and compare.
- Save as 'fake_v2.csv' and 'profile_fake_v2.xlsx'.

```python
# Step 6: Create improved fake dataset v2
np.random.seed(42)  # Reproducible

n_records = 500  # Same size

# Get frequencies from real for proportional sampling
year_freq = real_df['Year'].value_counts(normalize=True)  # Normalized frequencies
age_freq = real_df['Age'].value_counts(normalize=True)
sex_freq = real_df['Sex'].value_counts(normalize=True)

# Sample proportionally
real_years = sorted(real_df['Year'].unique())
fake_years = np.random.choice(real_years, n_records, p=year_freq.reindex(real_years, fill_value=0))  # Weighted
real_ages = real_df['Age'].unique()
fake_ages = np.random.choice(real_ages, n_records, p=age_freq.reindex(real_ages, fill_value=0))
real_sexes = real_df['Sex'].unique()
fake_sexes = np.random.choice(real_sexes, n_records, p=sex_freq.reindex(real_sexes, fill_value=0))

# For population: Fit normal dist per age-sex combo
pop_min, pop_max = real_df['Population'].min(), real_df['Population'].max()
fake_pop = []
for i in range(n_records):
    age = fake_ages[i]
    sex = fake_sexes[i]
    subset = real_df[(real_df['Age'] == age) & (real_df['Sex'] == sex)]['Population']  # Subset
    if len(subset) > 1:  # Need at least 2 for std
        mu, sigma = stats.norm.fit(subset)  # Fit normal dist
        pop = max(0, int(np.random.normal(mu, sigma)))  # Sample, ensure non-negative
    else:
        pop = np.random.randint(pop_min, pop_max + 1)  # Fallback
    fake_pop.append(pop)

# Create v2 DataFrame
fake_v2_df = pd.DataFrame({
    'Year': fake_years,
    'CountyFIPS': np.random.choice(real_df['CountyFIPS'].unique(), n_records),  # Sample county IDs
    'Sex': fake_sexes,
    'Age': fake_ages,
    'Population': fake_pop
})

# Save v2
output_path = os.path.join(output_dir, 'fake_v2.csv')
try:
    fake_v2_df.to_csv(output_path, index=False)  # Save
    print(f"Improved fake v2 saved as '{output_path}' with shape:", fake_v2_df.shape)  # Verification
    print(fake_v2_df.head())  # Preview
except PermissionError as e:
    print(f"PermissionError: {e}. Using fallback filename.")
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    fallback_path = os.path.join(output_dir, f'fake_v2_{timestamp}.csv')
    fake_v2_df.to_csv(fallback_path, index=False)
    print(f"Fallback: Saved as '{fallback_path}'")
```

```python
# Step 6: Profile fake_v2
fake_v2_stats = fake_v2_df.describe(include='all')  # Stats
print("Descriptive Statistics for Fake v2:\n", fake_v2_stats)  # Display

# Save profile
output_path = os.path.join(output_dir, 'profile_fake_v2.xlsx')
try:
    with pd.ExcelWriter(output_path, mode='w') as writer:
        fake_v2_stats.to_excel(writer, sheet_name='Descriptive_Stats')
except PermissionError as e:
    print(f"PermissionError: {e}. Using fallback filename.")
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    fallback_path = os.path.join(output_dir, f'profile_fake_v2_{timestamp}.xlsx')
    with pd.ExcelWriter(fallback_path, mode='w') as writer:
        fake_v2_stats.to_excel(writer, sheet_name='Descriptive_Stats')
    output_path = fallback_path

# Charts for v2
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

v2_pivot = fake_v2_df.pivot_table(values='Population', index='Age', columns='Sex', aggfunc='mean')
v2_pivot.plot(kind='bar', ax=axes[0], alpha=0.7, label='Fake v2')
try:
    with pd.ExcelWriter(output_path, mode='a', engine='openpyxl') as writer:
        v2_pivot.to_excel(writer, sheet_name='Chart1_Data')
except PermissionError as e:
    print(f"PermissionError: {e}. Using fallback filename.")
    fallback_path = os.path.join(output_dir, f'profile_fake_v2_{timestamp}.xlsx')
    with pd.ExcelWriter(fallback_path, mode='a', engine='openpyxl') as writer:
        v2_pivot.to_excel(writer, sheet_name='Chart1_Data')
    output_path = fallback_path

v2_yearly = fake_v2_df.groupby('Year')['Population'].sum()
v2_yearly.plot(kind='line', ax=axes[1], marker='^', label='Fake v2')
try:
    with pd.ExcelWriter(output_path, mode='a', engine='openpyxl') as writer:
        v2_yearly.to_excel(writer, sheet_name='Chart2_Data')
except PermissionError as e:
    print(f"PermissionError: {e}. Using fallback filename.")
    fallback_path = os.path.join(output_dir, f'profile_fake_v2_{timestamp}.xlsx')
    with pd.ExcelWriter(fallback_path, mode='a', engine='openpyxl') as writer:
        v2_yearly.to_excel(writer, sheet_name='Chart2_Data')
    output_path = fallback_path

plt.tight_layout()
plt.show()

# Final comparison: Real vs v1 vs v2
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Bar comparison
pivot_data.plot(kind='bar', ax=axes[0], alpha=0.7, label='Real')
fake_pivot.plot(kind='bar', ax=axes[0], alpha=0.5, label='Fake v1')
v2_pivot.plot(kind='bar', ax=axes[0], alpha=0.3, label='Fake v2')
axes[0].set_title('Comparison: Avg Population by Age/Sex')
axes[0].legend()

# Line comparison
yearly_total.plot(kind='line', ax=axes[1], marker='o', label='Real')
fake_yearly.plot(kind='line', ax=axes[1], marker='s', label='Fake v1')
v2_yearly.plot(kind='line', ax=axes[1], marker='^', label='Fake v2')
axes[1].set_title('Comparison: Total Population Over Years')
axes[1].legend()

plt.tight_layout()
plt.show()

# Numerical comparison for v2
v2_mean_diff = {}
for col in ['Year', 'Population']:
    real_mean = real_df[col].mean()
    v2_mean = fake_v2_df[col].mean()
    v2_mean_diff[col] = abs((v2_mean - real_mean) / real_mean * 100)
print("Mean % Differences (Real vs Fake v2):", v2_mean_diff)  # Should be lower than v1

# Analysis of improvements
print("\nAnalysis of Improvements:")
print("- Fake v1 uses uniform sampling, leading to unrealistic population distributions (e.g., high variance).")
print("- Fake v2 improves by:")
print("  * Sampling years, ages, and sexes with weights based on real data frequencies.")
print("  * Using normal distribution for population per age/sex combo, preserving real data variance.")
print("  * Result: Lower % mean differences in population (see above), closer alignment in charts.")
print("Further improvements could include: Using advanced synthetic data libraries (e.g., SDV) or modeling correlations.")

print("\nAll tasks complete! Zip the CSV/Excel files and this notebook for submission.")
```
